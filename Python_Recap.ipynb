```python
n = int(input("Enter a number: "))
Sum = 0
for i in range(0,n+1):
    Sum = Sum + i
    print("Current number = {} , Previous number = {}, Sum = {}".format(i, i-1, Sum))
```

```python
n = int(input("Enter a number: "))
num = n
rev = 0
while(n > 0):
    rem = n % 10
    rev = rev * 10 + rem
    n = n // 10
print(rev)
if num == rev:
    print("Same numbers")
else:
    print("Not same")
```

```python
import random
n = int(input("1. Rock\n2. Paper\n3. Scissor\nEnter your choice: "))
comp = random.randint(1,3)
print(comp)
if ((n == 1) & (comp == 1)):
    print("Tie")
elif((n == 1) & (comp == 2)):
    print("Computer Win")
elif((n == 1) & (comp == 3)):
    print("You Win")
elif((n == 2) & (comp == 1)):
    print("You Win")
elif((n == 2) & (comp == 2)):
    print("Tie")
elif((n == 2) & (comp == 3)):
    print("Computer Wins")
elif((n == 3) & (comp == 1)):
    print("Computer Win")
elif((n == 3) & (comp == 2)):
    print("You Win")
elif((n == 3) & (comp == 3)):
    print("Tie")
```

```python
n = int(input("Please enter the number of students: ")) 
a = open("myFile.csv",'a')
a.write("Srno, Student Name, English, Math, Science, Total, Percentage")
for i in range(1, n+1):
    name = input("Enter the student name: ")
    Eng = int(input("Please enter the English mark: ")) 
    Math = int(input("Please enter the Math mark: ")) 
    Sci = int(input("Please enter the Science mark: ")) 
    Total = Eng + Math + Sci
    Percen = (Total / 300) * 100
    a.write("\n{},{},{},{},{},{},{}%".format(i,name,Eng,Math,Sci,Total,Percen))
a.flush()
a.close()
```

```python
a = open("input.txt",'r')
lines = a.readlines()
a1 = lines[0].strip()
a2 = lines[1].strip()
a3 = lines[2].strip()
sum = 0
n = int(a2)
for i in range(1,n+1):
    sum = sum + i
prod = int(a1) * sum

b = open(a3,'w')
b.write("Sum = {}\nProduct = {}".format(sum,prod))
b.flush()
b.close()
a.flush()
a.close()
print("Done. Please check the file")
```
