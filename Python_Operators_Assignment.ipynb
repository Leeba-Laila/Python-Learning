```python
# Create two strings a and b with any integer values stored in it And perform Addition, Multiplication, Subtraction and a2 +b2

num1 = 5
num2 = 3
a = str(num1)
b = str(num2)
print("Addition : ",(a + b))
print("\nSubtraction : ",(a - b))
print("\nMultiplication : ",(a * b))
print("\nAddition of a^2 and b^2 : ",((a ** 2) + (b ** 2)))
```

```python
# Use Boolean operators (==, !=, >, >=, <, <= ) , between a and b and Display output

a = 5
b = 7
print("Result of == : ",(a == b))
print("Result of != : ",(a != b))
print("Result of > : ",(a > b))
print("Result of >= : ",(a >= b))
print("Result of < : ",(a < b))
print("Result of <= : ",(a <= b))
```

```python
# Create two strings x and y with any integer values stored in it And perform bitwise operations (&,^, | ,>>,<<)

num1 = 15
num2 = 10
a = str(num1)
b = str(num2)
print("Result of & : ",(a & b))
print("Result of ^ : ",(a ^ b))
print("Result of | : ",(a | b))
print("Result of >> : ",(a >> b))
print("Result of << : ",(a << b))
```
